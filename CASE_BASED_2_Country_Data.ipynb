```python
# Mengimpor library untuk membaca dateset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Mengimpor data 
df_country = pd.read_csv("Country-data.csv")
df_country
```

```python
# Mengapus Keterangan Kolom Country
list_Country = list(df_country.columns)
list_Country.remove('country')
```

```python
# Cek missing value

df_country.isnull().sum()
```

```python
# Rangkuman data

df_country.describe()
```

# EXPLORATORY DATA ANALYSIS

```python
# Menampilkan Boxplot
fig, ax = plt.subplots(nrows = 3, ncols = 3, figsize=(20, 10))
ax=ax.flatten()
int_cols= df_country.select_dtypes(exclude='object').columns
for x, i in enumerate(int_cols):
    sns.boxplot(x=df_country[i], ax=ax[x], color='#6ac493')
```

```python
# Menampilkan Corrplot

corr = df_country.corr()
mask = np.triu(np.ones_like(corr, dtype=np.bool))
f, ax = plt.subplots(figsize=(10, 10))
cmap = sns.light_palette('#6ac493', as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=None,center=0,square=True, annot=True, linewidths=.5, cbar_kws={"shrink": .9})
plt.show()
```

```python
# Menampilkan Pairplot

g = sns.pairplot(data=df_country,diag_kind='kde')
g.map_lower(sns.kdeplot, color="#6ac493")
```

# PREPROCESSING

```python
# Import Library
from sklearn.preprocessing import StandardScaler
df_country_scaled = StandardScaler().fit_transform(df_country.drop(['country'], axis=1))
```

```python
# Menampilkan Keterangan Baris dan Kolom

df_country_scaled.shape
```

```python
# Data Scaled

df_country_scaled = pd.DataFrame(df_country_scaled)
df_country_scaled.columns = list_Country
df_country_scaled
```

# K-MEANS CLUSTERING

```python
#Import Library

from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
# Pembuatan Model K-Means dan visualisasi Elbow

model = KMeans()
visualize = KElbowVisualizer(model, k=(1,10))
visualize.fit(df_country_scaled)
visualize.poof()
```

```python
# Model K-Means

model = KMeans(n_clusters=3, random_state=1)
model.fit(df_country_scaled)
```

```python
# Menampilkan Model Label

model.labels_
```

```python
# Pengklasterisasi

df_country['cluster_Id'] = model.labels_
df_country.head()
```

```python
# Scatterplot expor dan penghasilan

sns.scatterplot( x='exports', y='income', hue = 'cluster_Id' ,data = df_country, palette = 'Set2')
```

```python
# Scatterplot expor dan gdp

sns.scatterplot(x='exports', y='gdpp' , hue = 'cluster_Id' ,data = df_country, palette = 'Set2')
```

```python
# Scatterplot kelahiran dan kesehatan

sns.scatterplot(x='child_mort', y='health' , hue = 'cluster_Id' ,data = df_country, palette = 'Set2')
```

```python
# Scatterplot life expec dan kesehatan

sns.scatterplot(x='life_expec', y='health' , hue = 'cluster_Id' ,data = df_country, palette = 'Set2')
```

```python
# Menampilkan daftar negara perkluster

df_country.groupby(['cluster_Id','country']).mean()
```

```python
# Penginisiasian Kluster berdasarkan negara yang membutuhkan, mungkin membutuhkan, dan tidak membutuhkan bantuan

cat = {0:'Need Help',1:'Might need help',2:'No Help needed'}
df_country['cluster_Id']=df_country['cluster_Id'].map(cat)
```

```python
# Menampilkan peta dari negara yang membutuhkan, mungkin membutuhkan, dan tidak membutuhkan bantuan

px.choropleth(df_country, locationmode='country names', locations='country', color=df_country['cluster_Id'], title='Countries that need help',
              color_discrete_map={'Need Help':'#d13838','Might need help':'#fae56b','No Help needed':'#87bd82'} ,projection='equirectangular')
```

```python
# List negara yang tidak membutuhkan

no_need_help = df_country[df_country['cluster_Id']=='No Help needed']['country']
no_need_help
```

```python
# List negara yang mungkin membutuhkan

might_need_help = df_country[df_country['cluster_Id']=='Might need help']['country']
might_need_help
```
